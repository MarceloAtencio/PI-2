```python
import pandas as pd
```

ARCHIVO "BASE_OBS_AUTOS_2022.XLSX"

```python
df = pd.read_excel("../PI-2/01 - Datasets/Datasets Factores de Riesgo/base_obs_autos_2022.xlsx")
df.head(2)
```

```python
df.info()
```

```python
df = df[["Fecha_Inicial", "Dia_Habil", "Turno", "Zona", "Antiguedad", "Tipo_Distraccion", "Proteccion_Total"]]
df.head(2)
```

```python
df_autos = df.rename(columns={"Fecha_Inicial":"Fecha"})
```

```python
df_autos["Año"] = df_autos["Fecha"].dt.year
df_autos["Mes"] = df_autos["Fecha"].dt.month
df_autos = df_autos[["Año", "Mes", "Dia_Habil", "Turno", "Zona", "Antiguedad", "Tipo_Distraccion", "Proteccion_Total"]]
df_autos
```

```python
df_autos["Dia_Habil"] = df_autos["Dia_Habil"].replace({1:"Si", 2: "No"})
df_autos["Turno"] = df_autos["Turno"].replace({1:"Mañana", 2: "Tarde"})
df_autos["Zona"] = df_autos["Zona"].replace({1:"Centro", 2: "Norte", 3: "Oeste", 4: "Sur"})
df_autos["Antiguedad"] = df_autos["Antiguedad"].replace({1:"menor a 10 años", 2: "Mayor a 10 años", 9: "SD"})
df_autos["Tipo_Distraccion"] = df_autos["Tipo_Distraccion"].replace({1:"Celular", 2: "Fuma", 3: "Come",4: "Ninguno", 5: "Otro", 9: "SD"})
df_autos["Proteccion_Total"] = df_autos["Proteccion_Total"].replace({1:"Si", 2: "No"})
```

```python
df_autos.head(2)
```

```python
df_autos.to_csv("../PI-2/Datos_FR_Autos.csv", index=False)
```
