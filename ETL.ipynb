```python
import pandas as pd
```

ARCHIVO "HOMICIDIOS.XLSX"

```python
df = pd.read_excel("../PI-2/01 - Datasets/Datasets Siniestros Viales/homicidios.xlsx", sheet_name= "HECHOS")
df.head(2)
```

```python
df.info()
```

Solamente las columnas "Altura" y "Cruce" tienen datos nulos, pero eso se debe a que cuando el siniestro ocurre en un cruce de calle, la altura es dato vacío. La sumatoria de dichos valores NO nulos resulta en 654 valores. 

Por ende, de los 696 registros, hay 42 valores que en ambas columnas se encuentran vacios. A explorar rapidamente esos datos en el dataset, se denota que los mismos indican ocurrencia en AUTOPISTAS.

```python
df = df[["ID", "AAAA", "MM", "HH", "FECHA", "TIPO_DE_CALLE", "Calle", "Cruce", "COMUNA", "pos x", "pos y", "VICTIMA", "ACUSADO"]]
df["FECHA"] = df["FECHA"].dt.day_name()
df.head(2)
```

```python
df_homicidios = df.rename(columns={"AAAA":"Año", "MM":"Mes", "HH":"Hora", "FECHA":"Dia", "TIPO_DE_CALLE":"Lugar", "Calle":"Calle_ppal", "Cruce":"Calle_sec", "COMUNA": "Comuna", "pos x": "Longitud", "pos y": "Latitud", "VICTIMA": "Victima", "ACUSADO": "Acusado"})
df_homicidios.head(2)
```

```python
df_victimas = pd.read_excel("../PI-2/01 - Datasets/Datasets Siniestros Viales/homicidios.xlsx", sheet_name= "VICTIMAS")
df_victimas.head(2)
```

```python
df_victimas.info()
```

```python
df_victimas = df_victimas[["ID_hecho", "ROL", "SEXO", "EDAD", "FECHA_FALLECIMIENTO"]]
df_victimas = df_victimas.rename(columns= {"ID_hecho":"ID", "ROL":"Rol_vic", "SEXO":"Sexo", "EDAD": "Edad", "FECHA_FALLECIMIENTO": "Fecha_fallec" })
df_victimas.head(2)
```

```python
df_victimas["Sexo"] = df_victimas["Sexo"].replace({"MASCULINO":"M", "FEMENINO":"F"})
```

```python
# Intenta convertir la columna a fechas
df_victimas["Fecha_fallec"] = pd.to_datetime(df_victimas["Fecha_fallec"], errors = "coerce")
df_victimas.head()
```

```python
df_victimas["Fecha_fallec"] = df_victimas["Fecha_fallec"].apply(lambda x: "SI" if not pd.isnull(x) else "SD")
df_victimas
```

```python
# Merge basado en la columna común "ID"
df = pd.merge(df_homicidios, df_victimas, on="ID")
df.head()
```

```python
df_comunas = pd.read_excel("../PI-2/01 - Datasets/Datasets Siniestros Viales/comunas.xlsx")
df_comunas = df_comunas[["COMUNAS", "BARRIOS"]]
df_comunas = df_comunas.rename(columns = {"COMUNAS":"Comuna", "BARRIOS": "Barrios"})
df_comunas.head()
```

```python
df_completo = pd.merge(df, df_comunas, on="Comuna")
df_completo.info()
```

```python
df_completo.to_csv("../PI-2/Datos_Homicidios.csv", index=False)
```

ARCHIVO "LESIONES.XLSX"

```python
df = pd.read_excel("../PI-2/01 - Datasets/Datasets Siniestros Viales/lesiones.xlsx", sheet_name= "HECHOS")
df.head(2)
```

```python
df.info()
```

```python
df = df[["id", "franja_hora", "comuna", "tipo_calle", "acusado"]]
df.head(2)
```

```python
df_hechos = df.rename(columns={"id":"ID", "franja_hora":"Hora", "comuna":"Comuna", "tipo_calle":"Lugar", "acusado":"Acusado"})
df_hechos.head(2)
```

```python
df = pd.read_excel("../PI-2/01 - Datasets/Datasets Siniestros Viales/lesiones.xlsx", sheet_name= "VICTIMAS")
df.head(2)
```

```python
df.info()
```

```python
df = df[["ID hecho", "AAA", "MM", "VEHICULO_VICTIMA", "SEXO", "EDAD_VICTIMA", "GRAVEDAD"]]
df.head(2)
```

```python
df_victimas = df.rename(columns={"ID hecho":"ID", "AAA":"Año", "MM":"Mes", "VEHICULO_VICTIMA":"Victima", "SEXO":"Sexo", "EDAD_VICTIMA": "Edad", "GRAVEDAD": "Gravedad"})
df_victimas.head(2)
```

```python
df_victimas["Gravedad"].unique()
```

```python
df_victimas["Gravedad"] = df_victimas["Gravedad"].replace({"sd":"LEVE", "SD": "LEVE", "grave":"GRAVE"})
df_victimas["Gravedad"].unique()
```

```python
# Merge basado en la columna común "ID"
df_lesiones = pd.merge(df_victimas, df_hechos, on="ID")
df_lesiones.head()
```

```python
df_lesiones.info()
```

```python
df_lesiones.to_csv("../PI-2/Datos_Lesiones.csv", index=False)
```
